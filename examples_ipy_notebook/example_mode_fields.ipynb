```python
import sys
# use sys.path.insert to insert the path of liblp in the PYTHON path list
sys.path.insert(0,"../")
from liblp import besselmode
import numpy as np

from matplotlib import pyplot as plt


x = np.linspace(-2,2,500)   # in units of a
y = x 
```

In this example, the mode fields for V=4.5 are calculated. The u,w values are from the other example and can be calculated using the get_intersects function.

# LP01 mode

```python
m=0
u=1.958
w=4.052

m01 = besselmode(m, u, w, x, y)

plt.figure()
plt.imshow(m01, extent=(min(x), max(x), min(y), max(y)), clim=[-1,1], cmap='bwr')
plt.colorbar()
```
